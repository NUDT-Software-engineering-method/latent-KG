```python
def get_trgs_from_line(line):
    return line.strip().split(";")

# 查看测试集中有多少标签没有在训练集中出现
def trg_in_test_notin_train(data_dir):
    train_trg = os.path.join(data_dir, "train_trg.txt")
    valid_trg = os.path.join(data_dir, "valid_trg.txt")
    test_trg = os.path.join(data_dir, "test_trg.txt")
    train_trg_set = set([trg for tgs in open(train_trg, "r") for trg in get_trgs_from_line(tgs) ])
    valid_trg_set = set([trg for tgs in open(valid_trg, "r") for trg in get_trgs_from_line(tgs) ])
    test_trg_set = set([trg for tgs in open(test_trg, "r") for trg in get_trgs_from_line(tgs) ])
    print("Test not in train:", len(test_trg_set - train_trg_set))
    print("Valid not in train:", len(valid_trg_set - train_trg_set))
```

```python
trg_in_test_notin_train("../data/Weibo")
```

```python
trg_in_test_notin_train("../data/Twitter")
trg_in_test_notin_train("../data/StackExchange")
```

### Case Study
选取的规则为：
1. 在TRGKG中前五个全部预测正确
2. GATER前五个预测正确TRGKG中靠后或者预测错误
3. TAKG预测错误
打印规则例子为
原帖子：
目标关键词：
TAKG ：
GATER：
TRGKG：
空行

```python
from nltk.stem.porter import *
import copy
stemmer = PorterStemmer()
def stem_word_list(word_list):
    return [stemmer.stem(w.strip().lower()) for w in word_list]

def stem_str_list(str_list):
    stemmed_str_list = []
    for words_str in str_list:
        stemmed_word_list = stem_word_list(words_str.split(' '))
        stemmed_str_list.append(' '.join(stemmed_word_list))
    return stemmed_str_list
```

```python
# 读取数据
Pre_PATH = './caseStudy/twitter/'
golden_keyphrases = [line.strip().split(';') for line in open(Pre_PATH + 'test_trg.txt', 'r')]
source_posts = [line.strip() for line in open(Pre_PATH + 'test_src.txt', 'r')]
takg_predctions = [line.strip().split(';')[:5] for line in open(Pre_PATH + 'predictions_takg.txt', 'r')]
gater_predctions = [line.strip().split(';')[:5] for line in open(Pre_PATH + 'predictions_gater.txt', 'r')]
trgkg_predctions = [line.strip().split(';')[:5] for line in open(Pre_PATH + 'predictions_trgkg.txt', 'r')]
nums = len(golden_keyphrases) 
assert len(source_posts) == nums
assert len(trkg_predctions) == nums
assert len(gater_predctions) == nums
assert len(trgkg_predctions) == nums
# 转化成词根
golden_keyphrases_stream = [stem_str_list(keyphrases) for keyphrases in golden_keyphrases]
takg_keyphrases_stream =  [stem_str_list(keyphrases) for keyphrases in takg_predctions]
gater_keyphrases_stream = [stem_str_list(keyphrases) for keyphrases in gater_predctions]
trgkg_keyphrases_stream = [stem_str_list(keyphrases) for keyphrases in trgkg_predctions]
```

```python
#过滤数据 选取满足条件的数据
for index, line in enumerate(golden_keyphrases_stream):
    if len(line) < 2:
        continue;
    count = 0
    for key in line:
        if key in trgkg_keyphrases_stream[index]:
            count += 1
    if count != len(line): 
        continue;
    count_gate = 0
    count_takg = 0
    for key in line:
        if key in gater_keyphrases_stream[index]:
            count_gate += 1
    for key in line:
        if key in takg_keyphrases_stream[index]:
            count_takg += 1
    if count > count_gate and count > count_takg:
        print("ID:", index)
        print("Source Posts:", source_posts[index])
        print("Golden Keyphrases:", golden_keyphrases[index])
        print("TAKG:", takg_predctions[index])
        print("GATER:", gater_predctions[index])
        print("TRGKG:", trgkg_predctions[index])
        print()
```
