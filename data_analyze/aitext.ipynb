```python
from aitextgen.TokenDataset import TokenDataset
from aitextgen.tokenizers import train_tokenizer
from aitextgen.utils import GPT2ConfigCPU
from aitextgen import aitextgen
```

```python
file_name = "../data/Twitter/src_refs.txt"
```

```python
#ai = aitextgen(tokenizer_file=tokenizer_file, config=config)
ai = aitextgen(tf_gpt2="124M", to_gpu=True)
```

### 模型微调

```python
ai.train(file_name,
         line_by_line=True,
         from_cache=False,
         num_steps=2000,
         generate_every=500,
         save_every=500,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1, 
         )
```

### 模型加载

```python
ai = aitextgen(model_folder="trained_model", to_gpu=True)
```

### 生成文本

```python
ai.generate_one(batch_size=3,
            prompt="watched the packers game today at some friends house ! the packers are looking good ! super bowl baby ! go pack go ! <sep>",
            max_length=50,
            temperature=0.9,
            top_p=0.9)
```

```python
import os
from tqdm import tqdm
dataset = f"../data/Twitter"
train_src = os.path.join(dataset, "train_src.txt")
valid_src = os.path.join(dataset, "valid_src.txt")
test_src = os.path.join(dataset, "test_src.txt")
gen_list = []
num_file = sum([1 for i in open(valid_src, "r")])
with open(train_src, "r") as file:
    for idx, line in tqdm(enumerate(file), total=num_file):
        line = line.strip().replace("\n","")+"<sep>"
        result = ai.generate_one(batch_size=16, prompt=line, max_length=70, temperature=0.9, top_p=0.9)
        gen_texts = result.split("<sep>")
        if len(gen_texts)>1:
            gen_list.append(gen_texts[1])
        else:
            print(gen_texts)
            gen_list.append(gen_texts[-1])
```

```python
train_gen = os.path.join(dataset, "valid_gen.txt")
gen_file = open(train_gen, "w")
gen_file.write("\n".join(gen_list))
gen_file.close()
```
